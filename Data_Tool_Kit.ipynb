3)Explain the following terms
- The difference in np.array, np.as array and np.as any array
- The difference between Deep copy and shallow copy


np.array: Creates a new NumPy array from an array-like object (e.g., a list, tuple, or another array).
Behavior: Always creates a new array. The data is copied from the input object meaning changes to the original object will not affect the new array.

```python
# Create a NumPy array from a list
arr = np.array([1, 2, 3, 4])
arr
```

np.asarray: Converts an input to a NumPy array, but it does not copy the data if the input is already a NumPy array.
It is used to ensure the input is an array but avoids unnecessary copying.
Behavior: If the input is already a NumPy array, np.asarray returns a reference to it.
If the input is not an array, it creates a new array and copies the data.

```python
# Convert a list to a NumPy array
arr = np.asarray([1, 2, 3, 4])
arr
```

np.asanyarray :Similar to np.asarray, but it preserves subclasses of np.ndarray, such as np.matrix.
Behavior: Converts the input to an array but retains its subclass type if it is a subclass of np.ndarray.
This is useful when you want to ensure the result is an array but also need to maintain the original array’s subclass.

```python
# Create a matrix
matrix = np.matrix([[1, 2], [3, 4]])
matrix
```

Shallow Copy :Creates a new object but does not recursively copy the objects contained within it. Instead, it copies references to the objects.
Behavior: If the original object contains other objects
(e.g., a list of lists), the shallow copy will reference the same nested objects as the original. Changes to the nested objects in the copied object will be reflected in the original object.

```python
import copy

original_list = [[1, 2], [3, 4]]
shallow_copied_list = copy.copy(original_list)

# Modify the nested list in the shallow copy
shallow_copied_list[0][0] = 99

print(original_list )
print(shallow_copied_list)
```

Deep Copy Creates a new object and recursively copies all objects found within the original object, resulting in a completely independent copy.
Behavior: Changes to the nested objects in the deep copy will not affect the original object because all nested objects are also copied.

```python
import copy

original_list = [[1, 2], [3, 4]]
deep_copied_list = copy.deepcopy(original_list)

# Modify the nested list in the deep copy
deep_copied_list[0][0] = 99

print(original_list)

print(deep_copied_list)
```

```python
''' 14.Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')

# Generate random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.normal(loc=20, scale=5, size=len(date_range))
humidity = np.random.normal(loc=50, scale=10, size=len(date_range))

# Create DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})


df.set_index('Date', inplace=True)
df.head


#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
# right y axis for 'Humidity'

import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()
ax1.plot(df.index, df['Temperature'], 'b-', label='Temperature')

#b) Label the x-axis as 'Date'

ax1.set_xlabel('Date')
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(df.index, df['Humidity'], 'r-', label='Humidity')
ax1.set_ylabel('Temperature (°C)', color='b')
ax1.tick_params('y', colors='b')


ax2 = ax1.twinx()
ax2.plot(df.index, df['Humidity'], 'r-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='r')
ax2.tick_params('y', colors='r')



#c) Set the title of the plot as 'Temperature and Humidity Over Time'.
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm


np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

#a) Plot a histogram of the data with 30 bins.
plt.figure(figsize=(10, 6))
count, bins, patches = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

#b)Overlay a line plot representing the normal distribution's probability density function (PDF).

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2)

#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

'''16.Set the title of the plot as 'Histogram with PDF Overlay'.'''
#sol:
plt.title('Histogram with PDF Overlay')
plt.show()

```

```python
'''17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import seaborn as sns
```

```python
np.random.seed(1)
x = np.random.randn(100)
y = np.random.randn(100)

df = pd.DataFrame({
    'X': x,
    'Y': y
})

df['Quadrant'] = np.where((df['X'] >= 0) & (df['Y'] >= 0), 'Quadrant 1',
                  np.where((df['X'] < 0) & (df['Y'] >= 0), 'Quadrant 2',
                  np.where((df['X'] < 0) & (df['Y'] < 0), 'Quadrant 3', 'Quadrant 4')))


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='viridis', s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')


plt.legend(title='Quadrant')


plt.show()
```

```python
'''19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values,
label the axes, and set the title as 'Random Categorical Bar Chart'8
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.io import output_file

# Use output_notebook() to display plots directly in Jupyter notebooks
output_notebook()

# Generate random categorical data
np.random.seed(0)  # For reproducibility
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))
colors = ['#f07c7c', '#f0a25d', '#f0d23d', '#a2f07c', '#7cb9f0']

# Create a DataFrame
df = pd.DataFrame({
    'Category': categories,
    'Value': values,
    'Color': colors
})

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(df)

# Create a new figure
p = figure(title='Random Categorical Bar Chart')

# Add a bar renderer
p.vbar(x='Category', top='Value', width=0.5, color='Color', source=source, legend_field='Category')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Category'
p.yaxis.axis_label = 'Value'
p.xaxis.major_label_orientation = 'vertical'

# Set grid line colors and add a legend
p.grid.grid_line_color = 'gray'
p.legend.title = 'Categories'

# Show the plot
show(p)
```
